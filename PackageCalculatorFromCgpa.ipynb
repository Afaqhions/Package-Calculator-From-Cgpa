```python
# Importing all packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Dataset
dataset = pd.read_csv(r"C:\Users\User\Desktop\Projects\Machine Learning\Placement_1.csv")
dataset.head()
```

```python
# Validating dataset
dataset.isnull().sum()
```

```python
# Plotting Graph 
sns.scatterplot(x='cgpa',y='package',data=dataset)
plt.title("Graph between Cgpa & Pakage")
plt.show()
```

```python
# Splitting Train & Test Data
x = dataset[['cgpa']] #Cgpa Column
y = dataset[['package']] #Package Column
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=32)
```

```python
# Initiating Model 
model = LinearRegression()
```

```python
# Fitting the model to train 
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
# Evaluating 
# 1. Slope
print(f"Slope of Model: {model.coef_[0]}")
# Slope to check increment change in the pridiction against each unit
# ---------------------------------------------------------------------------------------------------------------------------
# Intercept
print(f"Intercpt:{model.intercept_}")
# ---------------------------------------------------------------------------------------------------------------------------
print(f"Mean Squared Error:{mean_squared_error(y_test, y_pred)}")
# ---------------------------------------------------------------------------------------------------------------------------
print(f"R^2 Error:{r2_score(y_test,y_pred)}")
```

```python
# Visualize the regression line
plt.scatter(x,y,color='blue', label="Actual Points")
plt.plot(x, model.predict(x), color='black', label='Regression Line')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
# Take input from user
cgpa_input = float(input("Enter your CGPA: "))

# Predict using the trained model
out = model.predict([[cgpa_input]])

# Print result
print(f"Your Predicted Package is: {out[0][0]} LPA")
```
